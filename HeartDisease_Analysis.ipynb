Draw Correlation Matrix

```python
plt.figure(figsize=(15,5))
sns.heatmap(df.corr(),annot=True)
```

Count of people having heart disease and not having heart disease

```python
df["target"].value_counts()
```

Count of  Male & Female in this Dataset and Gender Distribution According to The Target Variable

```python
df["sex"].value_counts()
```

```python
sns.countplot(x="sex",hue="target",data=df)
plt.xticks([0,1],["Female","Male"])
plt.legend(labels=["No disease","Disease"])
plt.show()
```

Check Age Distribution In The Dataset

```python
sns.distplot(df["age"],bins=20)
plt.show()
```

 Check Chest Pain Type

```python
df["cp"].value_counts()
```

```python
sns.displot(df["cp"])
plt.xticks([0.0,1.0,2.0,3.0],["Typical angina","Atypical angina","Non-anginal pain","Asymptomatic"])
plt.show()
```

Chest Pain Distribution As Per Target Variable

```python
sns.countplot(x="cp",hue="target",data=df)
plt.xticks([0,1,2,3],["Typical angina","Atypical angina","Non-anginal pain","Asymptomatic"])
plt.show()
```

Fasting Blood Sugar Distribution According To Target Variable

```python
sns.countplot(x="fbs",hue="target",data=df)
plt.legend(labels=["No disease","Disease"])
plt.show()
```

Resting Blood Pressure Distribution

```python
df["trestbps"].hist()
```

Comparing Resting Blood Pressure As Per Sex Column

```python
g=sns.FacetGrid(df,hue="sex")
g.map(sns.kdeplot,"trestbps",shade=True)
plt.legend(labels=["Female","Male"])
plt.show()
```

Plot Continuous Variables

```python
cate_val=[]
cont_val=[]
for col in df.columns:
  if df[col].nunique() <= 10:
    cate_val.append(col)
  else:
    cont_val.append(col)

cate_val
cont_val
```

```python
df.hist(cont_val)
plt.tight_layout()
plt.show()
```
